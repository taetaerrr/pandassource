```python
# cctv

cctv_df = pd.read_excel("./resources/CCTV.xlsx",skiprows=2,usecols="C, E:N")
cctv_df
```

```python
cctv_df.tail(2)
```

```python
# nan 

# cctv_df = cctv_df.dropna()

cctv_df.dropna(inplace=True)
```

```python
cctv_df.head(2)
```

```python
#   

cctv_df.drop(index=0,inplace=True)
```

```python
cctv_df.head(2)
```

```python
cctv_df.info()
```

```python
cctv_df.describe()
```

```python
#  
#  =>  , '2015   CCTV ' => 2015 
# 2020    : 2015  + 2016 + 2017 + 2018 + 2019 + 2020
cctv_df.columns
```

```python
#   : rename()

cctv_df.rename(columns={"":" ", '2015   CCTV ':'2015 '},inplace=True)
```

```python
#   : 2020    : 2015  + 2016 + 2017 + 2018 + 2019 + 2020

cctv_df["2020 "] = cctv_df["2015 "] + cctv_df["2016"] \
+ cctv_df["2017"] + cctv_df["2018"] + cctv_df["2019"]+cctv_df["2020"]
```

```python
cctv_df.head(3)
```

```python
cctv_df.drop(columns=['2020 '],inplace=True)
```

```python
#   : sum()

cctv_df["2020 "] = cctv_df[['2015 ', '2016', '2017', '2018', '2019', '2020']].sum(axis=1)
cctv_df.head(3)
```

```python
#    : '2015 ', '2016', '2017', '2018', '2019', '2020'

cctv_df.drop(columns=['2015 ', '2016', '2017', '2018', '2019', '2020'])
```

```python
cctv_df.drop(columns=cctv_df[cctv_df.columns.difference([' ','2021', '2022', '2023', '2024', '2020 '])], inplace=True)
```

```python
cctv_df.head(3)
```

```python
#   
# cctv  = '2021' + '2022' + '2023' + '2024'

cctv_df["cctv "] = cctv_df[['2021', '2022', '2023', '2024','2020 ']].sum(axis=1)
```

```python
cctv_df.head(3)
```

```python
#  cctv      

cctv_df[min(cctv_df['cctv ']) == cctv_df['cctv ']]

# cctv_df['cctv '].min()
```

```python
#  cctv      

cctv_df[max(cctv_df['cctv ']) == cctv_df['cctv ']]
```

```python
cctv_df.sort_values("cctv ").head(1)
cctv_df.sort_values("cctv ",ascending=False).head(1)
```
