#### Subset (일부분 추출)
- 행 기준
    - df.head(), df.tail(), df.sample()
- 열 기준

```python
import pandas as pd
import numpy as np

# df = pd.DataFrame({
#     "a":[4,5,6,6,np.nan],
#     "b":[7,8,np.nan,9,9],
#     "c":[10,11,12,np.nan,12],
# }, index=[1,2,3,4,5])

df = pd.DataFrame({
    "a":[4,5,6,6,np.nan],
    "b":[7,8,np.nan,9,9],
    "c":[10,11,12,np.nan,12],
}, index=list(range(1,6)))

df
```

```python
# range(시작값,종료값) : 종료값은 포함x
range(1,6)
list(range(1,6))
```

```python
# 일부분 행 추출 - df.head() / df.tail()
df.head()
```

```python
# 임의로 하나 추출

print(df.sample())
print(df.sample(2))
print(df.sample(frac=0.5))
```

```python
# b 컬럼에서 가장 큰 하나의 행

df.nlargest(1,"b")
```

```python
df.nsmallest(2, "a")
```

```python
# pip install seaborn

import seaborn as sns

df = sns.load_dataset("iris")
df.head(3)
```

```python
# 컬럼 추출

df["petal_length"]
df.petal_length
```

```python
df[["petal_length","sepal_width"]]
```

```python
columns = ["petal_length","sepal_width","species"]
df[columns].tail(3)
```

```python
# 행,열
# loc[], iloc[]
df.columns
# 모든행, 컬럼('sepal_length', 'sepal_width', 'petal_length') 추출
# df[['sepal_length', 'sepal_width', 'petal_length']]

# df.loc[행, 열] : 행, 열 표현시
df.loc[:,'sepal_length':'petal_length']

# 2~5행, 컬럼('sepal_length', 'sepal_width', 'petal_length') 추출
df.loc[2:5,'sepal_length':'petal_length']
```

```python
# iloc : 인덱스 번호 이용
# 뒤에 인덱스 값은 미포함

# 행은 10~19
df.iloc[10:20]

# 행 1,2,4 열 1,3
df.iloc[[1,2,4],[1,3]]
```

```python
df[df["petal_length"] > 3]
```

```python
# 행 : sepal_length > 5 큰 행만 추출하고, 컬럼은 sepal_length,sepal_width 추출

df.loc[df["sepal_length"] > 5 ,["sepal_length","sepal_width"]]
```
